```python
# importing libraries
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv('tweets_dataset.csv')
```

```python
# Prepare a list to store cleaned and structured data
structured_rows = []

# Iterate each row in the dataset
for index, row in df.iterrows():
    tweets = str(row['Tweets']).split('\n') # Treat the Tweet cell as a string and split by line breaks

    tweet = ' '.join(tweet.strip() for tweet in tweets if tweet.strip())  # Join all lines into a single sentence, removing extra spaces

  # Append the cleaned row
    structured_rows.append({
        'Date': row['Date'],
        'closing_price': row['closing_price'],
        'adj_close_price': row['adj_close_price'],
        'Tweet': tweet
    })
```

```python
# Convert to a new DataFrame
structured_df = pd.DataFrame(structured_rows)

# Save the structured dataset to a new CSV file
structured_df.to_csv('structured_tweets_dataset.csv', index=False)
structured_df.head(11)
```

```python
df2= pd.read_csv('data.csv')
df2.head(11)
```

```python
merged_df = structured_df.merge(df2, left_on=['closing_price', 'adj_close_price'], right_on=['close_price', 'adj_price'])
merged_df = merged_df.drop(columns=['close_price', 'adj_price'])
```

```python
# importing libraries
import re
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')
```

```python
# importing module
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
```

```python
# 1. Text Preprocessing
def preprocess(text):
    text = str(text).lower()                      # lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # remove URLs
    text = re.sub(r'\W', ' ', text)               # remove special characters
    text = re.sub(r'\s+', ' ', text).strip()      # remove extra spaces
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

merged_df['Clean_Tweet'] = merged_df['Tweet'].apply(preprocess)
```

```python
# Tokenization
merged_df['tokens'] = merged_df['Clean_Tweet'].apply(word_tokenize)
```

```python
# No of Token word
total_tokens = merged_df['tokens'].apply(len).sum()
print(f"Total number of tokens: {total_tokens}")
```

```python
# Stemming --
stemmer = PorterStemmer()
def stem_tokens(tokens):
    return [stemmer.stem(token) for token in tokens]
merged_df['stemmed_tokens'] = merged_df['tokens'].apply(stem_tokens)
```

```python
# No of Stemmed words
if 'stemmed_tokens' in merged_df.columns:
  total_stemmed_words = merged_df['stemmed_tokens'].apply(len).sum()
  print(f"Total number of stemmed words: {total_stemmed_words}")
else:
  print("Stemming was not performed.")
```

```python
# Lematization
lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]
merged_df['lemmatized_tokens'] = merged_df['tokens'].apply(lemmatize_tokens)
```

```python
# No of Lematize words
if 'lemmatized_tokens' in merged_df.columns:
  total_lemmatized_words = merged_df['lemmatized_tokens'].apply(len).sum()
  print(f"Total number of lemmatized words: {total_lemmatized_words}")
else:
  print("Lemmatization was not performed.")
```

```python
# Create corpus
corpus = merged_df['Clean_Tweet'].tolist()

print("\nFirst 5 elements of the corpus:")
print(corpus[:5])
```

```python
# 2. TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features = 500)  # Adjust max_features as needed
tfidf_matrix = vectorizer.fit_transform(merged_df['Clean_Tweet'])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Sentiment Analysis
analyzer = SentimentIntensityAnalyzer()

def get_sentiment_scores(tweet):
    scores = analyzer.polarity_scores(tweet)
    return scores['compound']  # use compound score for overall sentiment

merged_df['sentiment_score'] = merged_df['Clean_Tweet'].apply(get_sentiment_scores)
```

```python
# Categorize sentiment scores into labels
merged_df['sentiment_label'] = pd.cut(merged_df['sentiment_score'],
                                           bins=[-1, -0.05, 0.05, 1],
                                           labels=['negative', 'neutral', 'positive'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize sentiment distribution
plt.figure(figsize = (7, 4))
sns.countplot(x='sentiment_label', data=merged_df)
plt.title('Sentiment Distribution in Tweets')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')
plt.show()
```

```python
monthly_sentiment = merged_df.groupby('Month')['sentiment_score'].mean().reset_index()

# Plot average sentiment score per month
plt.figure(figsize=(10, 4))
sns.barplot( x='Month', y='sentiment_score',  data=monthly_sentiment)
plt.title('Average Sentiment Score per Month')
plt.xlabel('Month')
plt.ylabel('Average Sentiment Score')
plt.show()
```

```python
daily_sentiment_per_month = merged_df.groupby(['Month', 'Day'])['sentiment_score'].mean().reset_index()

# Plot daily average sentiment for each month
for month in daily_sentiment_per_month['Month'].unique():
    month_data = daily_sentiment_per_month[daily_sentiment_per_month['Month'] == month]
    plt.figure(figsize=(8,3))
    sns.barplot(data=month_data, x='Day', y='sentiment_score')
    plt.title(f'Tweet Sentiment Score for Month {month}')
    plt.xlabel('Day', fontsize=12)
    plt.ylabel('Average Sentiment Score')
    # plt.grid(True)
    plt.show()
```

```python
# Plot Closing Price vs. Sentiment Score as a scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(data=merged_df, x='sentiment_score', y='closing_price', alpha=0.5)
plt.title('Closing Price vs Tweet Sentiment Score')
plt.xlabel('Sentiment Score')
plt.ylabel('Closing Price')
plt.show()
```

```python
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plot distribution of positive
plt.figure(figsize=(7, 3))
sns.histplot(df['Positive'], bins =20, kde =True)
plt.title('Distribution of positive')
plt.xlabel('Positive')
plt.ylabel('Count')
plt.show()
```

```python
# plot distribution of Negative
plt.figure(figsize=(7, 3))
sns.histplot(df['Negative'], bins =20, kde =True)
plt.title('Distribution of Negative')
plt.xlabel('Negative')
plt.ylabel('Count')
plt.show()
```

```python
# plot distribution of Neutral
plt.figure(figsize=(8, 3))
sns.histplot(df['Neutral'], bins =20, kde =True)
plt.title('Distribution of Neutral')
plt.xlabel('Neutral')
plt.ylabel('Count')
plt.show()
```

```python
# Adjusted price vs closing price
plt.figure(figsize=(10,4))
plt.plot(df['adj_price'], label='Adjusted Price')
plt.plot(df['close_price'], label='Close Price', linestyle='--')
plt.title('Adjusted Price vs Close Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 4))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
